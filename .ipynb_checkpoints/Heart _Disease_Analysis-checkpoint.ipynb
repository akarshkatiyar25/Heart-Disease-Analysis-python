# Feature Engineering

## Creating Columns

```python
def encoded_hd(hd):
    if hd =='yes':
        return 1
    else:
        return 0

heart['heart_disease_encoded']= heart['heart_disease'].apply(encoded_hd)
```

```python
def categorize_age(age):
    if 21 <= age <= 40:
        return 'Young Adult'
    elif 41 <= age <= 60:
        return 'Adult'
    elif 61 <= age <= 80:
        return 'Senior'  
    else:
        return 'Unknown'
        
heart['age_group']= heart['age'].apply(categorize_age)
```

```python
def categorize_bp(bp):
    if bp < 120:
        return 'Normal'
    elif 120 <= bp <= 129:
        return 'Elevated'
    elif 130 <= bp <= 139:
        return 'Hypertension Stage 1'
    elif 140 <= bp <= 180:
        return 'Hypertension Stage 2'
    else:
        return 'Hypertensive Crisis'

heart['bp_category']= heart['trestbps'].apply(categorize_bp)
```

```python
def categorize_chol(chol):
    if chol < 200:
        return 'Desirable'
    elif 200 <= chol <= 239:
        return 'Borderline High'
    else:
        return 'High'

heart['chol_category']= heart['chol'].apply(categorize_chol)        
```

```python
def categorize_thalach(hr):
    if hr < 100:
        return 'Low'
    elif 100 <= hr <=120:
        return 'Below Average'
    elif 121 <= hr <=140:
        return 'Average'    
    elif 141 <= hr <=160:
        return 'Above Average'
    else:
        return 'Excellent'

heart['thalach_category'] = heart['thalach'].apply(categorize_thalach)
```

```python
def categorize_oldpeak(op):
    if op == 0:
        return 'Normal'
    elif 0 < op <= 1.0:
        return 'Mild'
    elif 1.0 < op <= 2.0:
        return 'Moderate'
    else:
        return 'Severe'

heart['oldpeak_category'] = heart['oldpeak'].apply(categorize_oldpeak)
```

```python
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi <= 24.9:
        return 'Normal Weight'
    elif 25 <= bmi <= 29.9:
        return 'Overweight'
    elif 30 <= bmi <= 34.9:
        return 'Obesity class I'
    elif 35 <= bmi <= 39.9:
        return 'Obesity class II'
    else:
        return 'Obesity class III'

heart['bmi_category'] = heart['bmi'].apply(categorize_bmi)
```

# Visualisation

## How does age vary by pressence of heart disease?

```python
heart_disease = heart[heart['heart_disease']=='yes']
age = heart_disease.groupby('age_group')['heart_disease'].value_counts().reset_index(name='count').rename(columns={'index':'age_group'})
age = age.sort_values('count', ascending=False)
age
```

```python
plt.figure(figsize=(8,6))

ax = sns.barplot(data = age, x='age_group', y = 'count')

for p in ax.patches:
    height= p.get_height()
    ax.text(
        p.get_x() + p.get_width() / 2,
        height,
        int(height),
        ha='center', va='bottom',
            fontsize=10, fontweight='bold',
    )

plt.xlabel("Age Group")
plt.ylabel("Heart Disease Patient")
plt.title("Age Group vs Heart Disease patient", fontweight='bold')


sns.despine()
plt.show()
```

The prevalence of heart disease increase with age, being higest among seniors (568) and lowest among young adult (163).

## Is there a differnce in cholestrol levels between patients with and without heart disease?

```python
chol_v_hd = heart.groupby("chol_category")["heart_disease"].value_counts().reset_index(name='count').rename(columns={'index':'age_group'})
chol_v_hd
```

```python
plt.figure(figsize=(8,6))

ax = sns.barplot(data = chol_v_hd , x='chol_category',y = 'count', hue = 'heart_disease')

for p in ax.patches:
    height= p.get_height()
    ax.text(
        p.get_x() + p.get_width() / 2,
        height,
        int(height),
        ha='center', va='bottom',
            fontsize=10, fontweight='bold',
    )

plt.xlabel("Chol Category")
plt.ylabel("Heart Disease Patient")
plt.title("Cholestrol vs Heart Disease", fontweight='bold')


sns.despine()
plt.show()
```

There is a strong association between cholestrol levels and heart disease. Patients with high cholestrol account for the largest number of heart disease cases (843), followed by those with desirable cholestrol(391) and borderline high cholestrol(168). Overall, the likelihood of heart disease is considerably higher among individuals with elevated cholestrol levels.
